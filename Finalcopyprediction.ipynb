```python
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    %matplotlib inline
    import math
```

```python
Delhihousing = pd.read_csv('D:\\Academic\\SEM-3\\S&S\\NPTEL\\Delhi.csv');
Banglorehousing = pd.read_csv('D:\\Academic\\SEM-3\\S&S\\NPTEL\\Bangalore.csv')
Kolkatahousing = pd.read_csv('D:\\Academic\\SEM-3\\S&S\\NPTEL\\Kolkata.csv')       
combined_df = pd.concat([ Banglorehousing,  Delhihousing, Kolkatahousing], ignore_index=True)
```

```python
X = combined_df['Area']
y = combined_df['Price']
```

```python
plt.scatter(X, y)
```

```python
covariance = combined_df['Price'].cov(combined_df['Area'])
variance_X = combined_df['Area'].var()
```

```python
slope = covariance / variance_X
intercept = combined_df['Price'].mean() - slope * combined_df['Area'].mean()
combined_df['Price_Prediction'] = slope * combined_df['Area'] + intercept
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(combined_df['Area'], combined_df['Price'], label='Actual Price', color='blue')
plt.plot(combined_df['Area'], combined_df['Price_Prediction'], label='Predicted Price', color='red')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression: Price Prediction based on Area')
plt.legend(loc='upper right', fontsize='medium')
plt.show()
```
